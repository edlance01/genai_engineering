```python
from openai import OpenAI
import sys

client = OpenAI()

# 1. Start a new conversation and store it
# Using the Responses API (responses.create)
stream = client.responses.create(
    model="gpt-4o",
    input=[{"role": "user", "content": "Explain quantum entanglement like I'm five."}],
    store=True,  # Tells OpenAI to save this interaction
    stream=True,  # Enables real-time token delivery
)

# 2. Handle the stream
print("Assistant: ", end="")
response_id = ""

for event in stream:
    # FIX: Nested ID access
    if event.type == "response.created":
        response_id = event.response.id  # The ID lives inside the response object

    # Printing the text chunks
    if event.type == "response.output_text.delta":
        print(event.delta, end="")
        sys.stdout.flush()

print(f"\n\nSaved Response ID: {response_id}")

# 3. Follow up using the stored state
# You don't need to re-send the "quantum entanglement" prompt!
follow_up = client.responses.create(
    model="gpt-4o",
    previous_response_id=response_id,  # Link to the previous turn
    input="Now explain it like I'm a PhD student.",
    stream=False,  # We can mix streaming and non-streaming in the same thread
)

print(f"\nFollow-up: {follow_up.output_text}")
```
