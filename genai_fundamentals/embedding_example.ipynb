### With SentenceTransformer

```python
from sentence_transformers import SentenceTransformer

# Load the model
model = SentenceTransformer("all-MiniLM-L6-v2")
sentence = "AI is cool"

# 1. Get the Sentence Embedding (1 vector)
sentence_vector = model.encode(sentence)

# 2. Get the Token Embeddings (Individual vectors for 'AI', 'is', 'cool')
# We use output_value="token_embeddings"
token_embeddings = model.encode([sentence], output_value="token_embeddings")[0]

print(f"Sentence vector shape: {sentence_vector.shape}")  # (384,)
print(f"Token embeddings shape: {token_embeddings.shape}")  # (5, 384)
# Note: Shape is 5 because of [CLS], AI, is, cool, [SEP]
```
